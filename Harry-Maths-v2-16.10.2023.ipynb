```python
'''
Maths problem - 16th October 2023

Creat as many 2x2 square solutions as possible where a, b, c and d must be unique digits between 1-9

a b
c d

where: ab + ac + cd + bd = 200

therefore:
(a*10 + b) + (a*10 + c) + (c*10 + d) + (b*10 + d) = 200
'''
```

```python
# Here we import the library which provides the random number generator tools
import random

# Do we want debug output too? 0 - no / 1 - yes
debug = 0
```

```python
def find_a_solution():
    '''
    This function is used to find a solution to the problem
    '''
    # re-initialise variables
    if debug: print('initilising variables')
    a = 0
    b = 0
    c = 0
    d = 0
    test = 0

    # loop until you find a solution which = 200
    while test != 200:

        a = random.randint(1,9)
        b = random.randint(1,9)
        while b == a:
            b = random.randint(1,9)
        c = random.randint(1,9)
        while c == a or c == b:
            c = random.randint(1,9)
        d = random.randint(1,9)
        while d == a or d == b or d == c:
            d = random.randint(1,9)

        # check does this combination = 200?
        test = (a*10 + b) + (a*10 + c) + (c*10 + d) + (b*10 + d)
        
    # while loop exits so we can set current variables to our new answer 
    return([a, b, c, d])
```

```python
def print_solutions(solutions):
    '''
    This function is used to print out all the solutions nicely
    '''
    
    for pos, digit in enumerate(solutions):
        print("answer ", pos+1)
        print('', digit[0], digit[1], "\n", digit[2], digit[3])
        print('Test: ', (digit[0]*10 + digit[1]) + (digit[0]*10 + digit[2]) + (digit[2]*10 + digit[3]) + (digit[1]*10 + digit[3]), '\n')
```

```python
def main():
    '''
    This is the main function which creates the main loop operation to find a pre-defined number of solutions
    '''

    # Initialise some variables which we're going to use
    solutions = []
    num_solutions = 22
    attempts = 0

    while len(solutions) < num_solutions:

        # re-initialise variables
        answer = []
        solution = []
        
        if attempts > 100:
            exit

        # get an answer to the problem
        answer = find_a_solution()
        if debug: print('new answer: ', answer)
    
        # The first answer needs adding to the solutions array, exit and return to start of loop
        if not solutions:
            solutions.append(answer)
            if debug: print('our first solution: ', solutions, '\n')
            # Now flip back to start of the while loop for next solution
            continue

        # check whether there is already an array within our solutions array that matches
        result = any(answer in solutions for solution in solutions)
        if debug: print('Matching?: ', result)
        if not result:
            solutions.append(answer)
        if result:
            attempts =+ 1
            if debug: print('attempts: ', attempts)
            
        # show what solutions we have at this point
        if debug: print('our current solutions: ', solutions, '\n')

    print_solutions(solutions)
```
